```python
import pandas as pd
import matplotlib.pyplot as plt


##load datafile
df = pd.read_csv('UberDatasetCleaned.csv')
```

```python
df.head()
```

```python
df.info
```

```python
df.shape
df ['CATEGORY'] . value_counts(),
df ['PURPOSE'] .value_counts(dropna=False),
df ['START'] .value_counts().head(10)
```

```python
speed_per_day = df.groupby('day')['speed'].mean()
print(speed_per_day)
```

```python
 ##plot 
plt.figure(figsize=(10,5))
plt.bar(speed_per_day.index,
        speed_per_day.values),
plt.title('speed per day'),
plt.xlabel('day'),
plt.ylabel('speed'),
plt.tight_layout()
plt.show()
```

```python
speed_per_month= df .groupby ('month')['speed'] .mean ()
print (speed_per_month)
```

```python
# plotting a line chart for speed_per_month
plt.figure (figsize=(10,5))
plt.plot(speed_per_month .index,
        speed_per_month.values),
plt.title ('speed_per_month'),
plt.xlabel ('month'),
plt.ylabel('speed')
plt.tight_layout()
plt.show()
```

```python
average_speed =df['speed'].mean()
print('Average Speed')
print(average_speed)
```

```python
average_speed_by_day = df.groupby('day')['speed'].mean()
print('Average Speed By Day', average_speed_by_day)
```

```python
average_speed_day_time=df.groupby('day_time')['speed'].mean()
print('Average Speed On Daytime',average_speed_day_time)
```
